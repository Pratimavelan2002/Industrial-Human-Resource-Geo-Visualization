```python
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

# ------------------------------------------------------
# 1. Load Data
# ------------------------------------------------------
df = pd.read_csv(
    "C:/Users/BHUVANESWARI/OneDrive/Desktop/guvi project/Industrial Human Resource Geo-Visualization/data/merged_workers_nlp.csv"
)

# Ensure the required columns exist
assert "clean_text" in df.columns, "clean_text column missing!"
assert "industry_category" in df.columns, "industry_category missing!"

X = df["clean_text"]
y = df["industry_category"]

# ------------------------------------------------------
# 2. Create Model Pipeline
# ------------------------------------------------------
model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LinearSVC())
])

# ------------------------------------------------------
# 3. Train Model
# ------------------------------------------------------
model.fit(X, y)

# ------------------------------------------------------
# 4. Save Pipeline (Vectorizer + Model)
# ------------------------------------------------------
with open("nlp_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("✔ Training complete!")
print("✔ Model saved as nlp_model.pkl")
```
